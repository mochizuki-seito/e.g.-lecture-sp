#　分析結果・考察レポート　

### (homerun.ipynb とbatting_style_analysis.ipynbから)

プロ野球データ分析レポート（OPS・年俸・打撃スタイルの比較）
# 1. 研究目的
本研究では、プロ野球選手 3970 件のデータを分析し、以下の疑問を検証
* 年齢や血液型は OPS（打撃総合指標）や年俸と関係するのか
* パワーヒッター(PH)とアベレージヒッター(AH)では OPS と年俸に差があるのか
* 打撃スタイル(PH/AH)と年齢・血液型の関係は存在するのか

→球団の編成や補強にどんな示唆が得られるかを考える

これからの分析においてー
※本分析では、同一選手の年度ごとのデータを独立したサンプルとして扱っています。
これは、年度ごとにコンディションや環境が異なり、パフォーマンスが変動するためです。
ただし、同一選手の影響が重複して含まれている点には留意が必要

# 2. 全体分析（全 3970 件）

## 2-1. 年齢 × OPS の関係

- 相関係数：0.1097（無相関）

結論：
年齢だけでは OPS の高低を説明できない。
若手でも高OPSの選手は存在し、ベテランでも高パフォーマンスを維持している選手が多い。


## 2-2. 血液型 × OPS
ANOVA：有意差あり（p < 0.05）
p = 0.000
平均OPS値: A型(0.649), AB型(0.685), B型(0.666), O型(0.665)
最大差は0.036（約5.5%）だが、この差は...
- 選手個人の日々の変動より小さい
- チーム成績への実質的影響は限定的
- 他の要因（打撃スタイル等）の方が影響大と考えられる

## 2-3. 血液型 × 年俸
ANOVA：有意差あり（p < 0.05）
p = 0.000
平均年俸: A型(739万円), AB型(1574万円), B型(580万円), O型(1044万円)
AB型と他の血液型で差が大きいが...
- サンプルの偏りによる可能性が高い（AB型の割合は全体の約10%）
O     1300人
A     1284人
B      847人
AB     273人
- 年俸は実績・経験・ポジションなど多要素で決定
- 同じ血液型内でも大きなばらつきあり



## 2-4 三振率 (K_Rate) と OPS の関係

相関係数: -0.1185
ほとんど相関がない

結論：
三振率とOPSにはほとんど関係はない

## 2-5ホームラン率 (HR_Rate) と OPS の関係
正の相関がある
相関係数: 0.6911
p値: 0.0000(統計的に有意:p<0.05)

結論：
ホームラン率が高いとOPSが高い



# 3. 打撃スタイル別分析（パワー vs アベレージ）

パワーヒッター：29名

アベレージヒッター：36名

* 条件
- パワーヒッター: 三振率≥ 約0.23  → 三振が多い & HR≥15本 & 打率<.280 & 試合≥100
- アベレージヒッター: 三振率≤ 約0.15→ 三振が少ない & 打率≥.280 & HR<15本 & 試合≥100

## 3-1. OPS の比較
指標	パワー	アベレージ
平均OPS	0.792	0.751
t検定	有意差あり
p値: 0.0000（p < 0.05）

結論：
パワーヒッターは OPS が明確に高い。

解釈：

OPSは長打力に強く依存
単打主体の打者は出塁率をかなり高めないとパワー系に届かない

## 3-2. 年俸の比較
指標	パワー	アベレージ
平均年俸	1億6990万円	1億2283万円
t検定	有意差ありp = 0.0074	(p<0.05)

結論：
球団はパワーヒッターにより高い価値をつけている。

「希少性 × 長打力 × OPS」が高い市場価値に直結。

## 3-3. コストパフォーマンス（OPS / 1000万円、1000万円あたりのOPS）

パワーヒッター: 0.2266
アベレージヒッター: 0.1979

結論：
パワーヒッターは年俸が高くても、それ以上に価値（OPS）を出している。
→ 投資効率が統計的に有意に良い。

## 3-4. 年齢比較
指標	パワー	アベレージ
平均年齢	27.0歳	28.9歳
p値	0.0007  (p < 0.05)（有意差あり）	

結論：
アベレージヒッターの方が平均して年齢が高い。

解釈：

若手は三振率が高く、パワー型になりやすい

ベテランはコンタクト能力が伸びやすい


## 3-5. 年齢 × OPS の相関（タイプ別）

パワー：-0.117

アベレージ：-0.061

どちらも ほぼ相関なし

結論：
スタイルに関係なく、OPSは年齢によってほとんど変化しない。
→ 衰えは OPS では測りにくい。

## 3-6. 血液型 × 打撃スタイル

血液型分布：有意差なしp = 0.33（p > 0.05）

結論: 血液型の分布に統計的に有意な差なし
→ 打撃スタイルと血液型に明確な関係はない

# 4. 総合結論
① OPS
→ パワーヒッターが有意に高い
② 年俸
→ パワーヒッターの方が高い
③ コストパフォーマンス
→ パワーヒッターが優秀
④ 年齢
アベレージヒッターはややベテラン寄り
しかし OPS との相関は弱い

⑤ 血液型
→ 統計的検定では一部有意差が出たが、差の大きさはほとんど無視できるレベル
→ 打撃タイプとの関連も検出されず（p = 0.33）
→ よって球団運営において血液型を考慮する意義はないと結論づけられる


# 5. 球団編成への示唆と考察
① 補強方針
OPS向上を狙うなら“パワーヒッター優先”が合理的。
長打力は市場価値に合った実力を持ち、コスパも良い。

② 育成方針（打撃のみを考えた時）
若手 → パワー系のポテンシャルを重視
ベテラン → コンテクト能力強化で価値維持

③ 契約更改

年齢で単純に評価しない

OPS を主軸とした評価が効果的ではないか

④ 募集・育成における「血液型」は不要
科学的根拠も分析結果もどちらも否定。

⑤クリーンナップ（定石である３、４、５番）にパワーヒッターを置くのは理にかなっていると考えられる

### *しかし、チーム内での役割や球場の特性、守備などの打撃以外の能力などを考慮する必要がある
